# READING JSON

```python
import json
fd = open("record.json",'r')
r = fd.read()
fd.close()

record = json.loads(r)
```

# ADD NEW ITEMS TO THE INVENTORY

```python
prod_id = str(input("Enter product id:"))
name = str(input("Enter name:"))
expirydate=str(input("enter expirydate:"))
mfd=str(input("enter mfd:"))
sellername = str(input("Enter sellername:"))
if prod_id == record[prod_id]:
    new_quantity = int(input("Enter quantity:"))
    quantity= int(new_quantity + int(record[prod_id]["quantity"])
quantity = int(input("Enter quantity:"))
prize = int(input("Enter price:"))

record[prod_id] = {'name': name, 'expirydate': expirydate, 'mfd':mfd, 'sellername':sellername,  'quantity': quantity, 'prize': prize,}

js = json.dumps(record)

fd = open("record.json",'w')
fd.write(js)
fd.close()
```
